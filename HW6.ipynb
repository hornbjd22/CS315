```python
import random
import time
from statistics import mean

def quickselect(A, i):
    """Return the i-th smallest element of A (1-based i)."""
    if len(A) == 1:
        return A[0]

    pivot = random.choice(A)
    lows = [x for x in A if x < pivot]
    highs = [x for x in A if x > pivot]
    pivots = [x for x in A if x == pivot]

    if i <= len(lows):
        return quickselect(lows, i)
    elif i <= len(lows) + len(pivots):
        return pivot
    else:
        return quickselect(highs, i - len(lows) - len(pivots))


# Quick test harness
def time_quickselect(n, trials=5):
    times = []
    for _ in range(trials):
        A = list(range(n))                      # sorted
        random.shuffle(A)                       # random permutation
        k = random.randint(1, n)
        t0 = time.perf_counter()
        _ = quickselect(A, k)
        t1 = time.perf_counter()
        times.append(t1 - t0)
    return mean(times)

# Example runs (small n)
if __name__ == "__main__":
    for n in [100, 1000, 5000, 10000]:
        print(n, time_quickselect(n, trials=3))
```
